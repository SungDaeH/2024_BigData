| 함수 | 기능 |
|---------|-----------|
| query() | 행 추출 |
| df[] |  열(변수) 추출 |
| sort_values() | 정렬 |
| groupby() | 집단별로 나누기 |

| merge() | 데이터 합치기(열) |

```python
import pandas as pd
import numpy as np
exam = pd.read_csv("exam.csv")
exam.head()
```

```python
# exam nclass가 1인 경우만 추출

exam.query("nclass == 1")
```

```python
# exam nclass가 2인 경우만 추출
exam.query('nclass == 2')
```

```python
# exam nclass가 3인 경우만 추출
exam.query("nclass == 3")
```

```python
# exam nclass가 4인 경우만 추출
exam.query("nclass == 4")
```

```python
# exam nclass가 5인 경우만 추출
exam.query("nclass==5")
```

```python
# 문제 수학점수가 50점 초과한 경우만 출력
exam.query("math > 50")
```

```python
# 수학점수가 50점 미만인 경우만 출력
exam.query("math < 50")
```

```python
# 1반이면서 수학점수가 50점 이상인 경우만 추출
exam.query("nclass==1 & math >= 50")
```

```python
# 2반이면서 영어점수가 80점 이상인 경우만 추출
exam.query("nclass == 2 & english >= 80")
```

```python
# 수학점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query("math >= 90 | english >= 90")
```

```python
# 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query("english < 90 | science < 50")
```

```python
# # 1,3,5반에 해당하는 값 출력
# exam.query("nclass == 1 | nclass == 3 | nclass == 5")
exam.query('nclass in [1,3,5]')
```

```python
# nclass가 1인 행을 추출해서 nclass1 할당
# nclass가 2인 행을 추출해서 ncalss2 할당
nclass1 = exam.query("nclass == 1")
nclass2 = exam.query("nclass == 2")
```

```python
# 1반 수학 점수 평균
nclass1["math"].mean()
```

```python
# 2반 수학 점수 평균
nclass2["math"].mean()
```

```python
nclass1["english"].mean() # 1반 영어 점수 평균
```

```python
nclass2["english"].mean() # 2반 영어 점수 평균
```
