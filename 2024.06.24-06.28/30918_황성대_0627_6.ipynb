```python
# 예상문제1
# 꺽은선 그래프 그리기
# 라이브러리 불러오기 및 별명 저장하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc("font", family="Malgun Gothic")

# x, y축 데이터 저장
sales = [1,5,7,3,7]
month = ['mar', 'apr', 'may', 'jun', 'jul']

# 그래프의 해상도 설정하기 dpi.figure
plt.figure(dpi = 100)

# 제목 title
plt.title('월별 판매 실적')

# 꺽은선형 그래프 그리기 .plot .color
# 그래프 범례 .label
plt.plot(month, sales, color = 'm', label='단위:개수')

# 범례보이게 설정 위치 .lower right
plt.legend(loc = 'lower right')
#x축 y축 xlabel, ylabel
plt.xlabel('월')
plt.ylabel('매출')

# 그래프 보이기
plt.show()
```

```python
# 예상문제 5
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font', family = 'Malgun Gothic')

# 데이터 만들기

height = []
weight = []
for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
    
# 제목 : 키와 몸무게의 상관관계
plt.title('키와 몸무게의 상관관계')

# 산점도 그래프 그리기
plt.scatter(height, weight, c = height, cmap='RdPu')

# 축제목

plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')
```

```python
# 예상문제 5
# 라이브러리 불러오기 및 별명 저장하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font', family = 'Malgun Gothic')
import random

# 데이터 만들기
height, weight, score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    

# 제목 : 키와 몸무게의 상관관계
plt.title("키와 몸무게의 상관관계")

# 산점도 그래프 그리기
plt.scatter(height, weight, s = score, c = score, cmap = 'rainbow', alpha = 0.7)

# 축제목
plt.colorbar(label='체육 실기 점수')
plt.show()
```

```python
# 예상문제 7
# 라이브러리 불러오기 및 별명 저장하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font', family = "Malgun Gothic")

# 데이터
singer = ["A","B","C","D","E"]
week1 = [42, 58, 19, 92, 84]
week2 = [53, 52, 48, 98, 73]
# 마이너스 데이터 만들기
for i in range(len(week1)):
    week1[i] = -week1[i]
    
# 마이너스 기호 표시
plt.rcParams['axes.unicode_minus'] = False

# 제목
plt.title("오디션 프로그램 득표 현황")

# 가로차트 그리기
plt.barh(singer, week1, label='첫째주', color='hotpink')
plt.barh(singer, week2, label = '둘째주', color = 'royalblue')

plt.legend()
plt.show()
```
