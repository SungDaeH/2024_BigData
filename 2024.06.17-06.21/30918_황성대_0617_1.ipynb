## 원형 차트 그리기

```python
#라이브러리 불러오가 맟 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트
plt.rc('font', family = 'Malgun Gothic')
#data 혈액형 수
b_type = [25,19,37,11]
b_name = ["A형","B형","O형","AB형"]
#제목 : 혈액형 분포
plt.title("혈액형 분포")
# 원형차트 그리는 명령어 pie 범례 labels = list[] autopct = ''

plt.pie(b_type, labels=b_name, autopct = '%1.1f%%')
plt.show()
```
